```python
from plotly.offline import iplot
import plotly.graph_objs as go

def plot_column_counts(column_name):
    import preprocesamiento_datos_train
    import preprocesamiento_datos_test
    from sklearn.preprocessing import StandardScaler
    X_train, y_train, RANDOM_STATE, scaler = preprocesamiento_datos_train.preprocesamiento(False,[], False)
    # Obtener los datos preprocesados
    X_test, y_test, RANDOM_STATE = preprocesamiento_datos_test.preprocesamiento(False,[], False, scaler)
    
    # Contamos el nmero de ocurrencias de '1' para la columna especificada en train y test
    train_counts = X_train[column_name].sum()
    test_counts = X_test[column_name].sum()
    
    # Obtenemos el nmero total de muestras en cada conjunto de datos
    total_train_samples = len(X_train)
    total_test_samples = len(X_test)
    
    # Normalizamos los recuentos dividindolos por el nmero total de muestras en cada conjunto
    train_counts_normalized = train_counts / total_train_samples
    test_counts_normalized = test_counts / total_test_samples
    
    # Creamos los datos para los grficos de barras
    trace1 = go.Bar(
        x=["Train"],
        y=[train_counts_normalized],
        name='Train'
    )
    
    trace2 = go.Bar(
        x=["Test"],
        y=[test_counts_normalized],
        name='Test'
    )
    
    # Definimos el diseo del grfico
    layout = dict(
        width=800,
        height=600,
        title=dict(
            text="Proporcin de {} en Train y Test".format(column_name)
        ),
        xaxis=dict(
            title="Conjunto de Datos"
        ),
        yaxis=dict(
            title="Proporcin de '1'"
        ),
        # Color de fondo de la figura
        paper_bgcolor="LightSteelBlue",
        # Color de fondo del grfico
        plot_bgcolor="#F5EDA0",
    )
    
    # Creamos la figura y dibujamos
    fig = go.Figure(data=[trace1, trace2], layout=layout)
    
    return fig

# Llama a la funcin para generar el grfico
fig = plot_column_counts("nationality_Germany")  # Puedes cambiar el nombre de la columna aqu

# Muestra el grfico
iplot(fig)
```
